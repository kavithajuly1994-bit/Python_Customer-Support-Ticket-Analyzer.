# **Step 2: Add More Tickets**

* Ask the user:  How many new tickets do you want to add?
* For each new ticket, collect:
 * Customer Name
 * Issue Description
 * Priority (High / Medium / Low)

**Requirements:**

* Ticket numbers must auto-increment starting from 11.
* Validate priority (accept only High, Medium, Low).
* Append new data to ticket_data.


```python
# Ask user how many tickets to add
num_tickets = min(int(input("Enter number of tickets (max 5): ")), 5)

# Start ticket number from next value
next_ticket_no = max(ticket_data['Ticket_No']) +1

for i in range(num_tickets):
    print(f"\nEnter details for Ticket {next_ticket_no}:")

    name = input("Customer Name: ")
    issue = input("Issue Description: ")

    # Priority validation
    while True:
        priority = input("Priority (High / Medium / Low): ").capitalize()
        if priority in ['High', 'Medium', 'Low']:
            break
        else:
            print("Invalid input! Please enter High, Medium, or Low.")

    # Append data
    ticket_data['Ticket_No'].append(next_ticket_no)
    ticket_data['Customer_Name'].append(name)
    ticket_data['Issue_Description'].append(issue)
    ticket_data['Priority'].append(priority)

    next_ticket_no += 1

# Print updated data
print("\nUpdated Ticket Data:")
for i in range(len(ticket_data['Ticket_No'])):
    print(f"{ticket_data['Ticket_No'][i]}"
          f"{ticket_data['Customer_Name'][i]}"
          f"{ticket_data['Issue_Description'][i]}"
          f"{ticket_data['Priority'][i]}")
```

# Step 3: Text Cleaning for Issue Descriptions

* Clean all issue descriptions by applying:
     * Remove punctuation (.,!?-)
     * Convert multiple spaces → single space
     * Remove leading/trailing spaces
     * Convert text to lowercase
     * Replace slang/shorthand (example: "ok" → "okay")
     
**Hints**:
 Use .replace(), .split(), ' '.join(), .strip(), .lower().


```python
for i in range(len(ticket_data['Issue_Description'])):

    text = ticket_data['Issue_Description'][i]

# Step 1: lowercase
text = text.lower()

# Step 2: replace slang
text = text.replace("ok", "okay")

# Step 3: remove punctuation
for p in ".,!?-":
    text = text.replace(p, "")

# Step 4: remove extra spaces
text = ' '.join(text.split())

# Step 5: strip spaces
text = text.strip()


print("Cleaned Issue Descriptions:")

for issue in ticket_data['Issue_Description']:
    print(issue)



```

# **Step 4: Keyword-Based Issue Insights**
Create a function:
def count_tickets_with_word(word):

**Function Requirements:**
* Case-insensitive search
* Returns how many ticket descriptions contain the given word


**Use this function to print:**
* Number of tickets containing "poor"
* Number of tickets containing "good"
* Number of tickets containing "slow"
* Number of tickets containing "excellent"


```python
def count_tickets_with_word(word):
    count = 0

    for issue in ticket_data['Issue_Description']:
        # Case-insensitive check
        if word.lower() in issue.lower():
            count += 1

    return count

print("Tickets with 'poor':", count_tickets_with_word("poor"))
print("Tickets with 'good':", count_tickets_with_word("good"))
print("Tickets with 'slow':", count_tickets_with_word("slow"))
print("Tickets with 'excellent':", count_tickets_with_word("excellent"))
```

**Priority Analysis**
Compute:
* Number of High-priority tickets
* Number of Medium priority tickets
* Number of Low-Priority tickets


```python
high = 0
medium = 0
low = 0

for p in ticket_data['Priority']:
    if p == "High":
        high += 1
    elif p == "Medium":
        medium += 1
    elif p == "Low":
        low += 1

print("\nPriority Analysis:")
print("High Priority Tickets  :", high)
print("Medium Priority Tickets:", medium)
print("Low Priority Tickets   :", low)
```

**Find the Ticket With the Longest Issue Description**
Based on word count, print:
* Ticket number
* Customer name
* Cleaned issue text
* Word count


```python
# Find longest issue description
max_words = 0
index = 0

for i in range(len(ticket_data['Issue_Description'])):
    words = ticket_data['Issue_Description'][i].split()
    word_count = len(words)

    if word_count > max_words:
        max_words = word_count
        index = i

# Print result
print("\nTicket with Longest Issue Description:")
print("Ticket No   :", ticket_data['Ticket_No'][index])
print("Customer    :", ticket_data['Customer_Name'][index])
print("Issue       :", ticket_data['Issue_Description'][index])
print("Word Count  :", max_words)
```

**Extract Unique Words Used**
Generate a set of all unique words across all issue descriptions.

**Show**:
* Count of unique words
* The word list (sorted)


```python
unique_words = set()

for issue in ticket_data['Issue_Description']:
    words = issue.split()
    for word in words:
        unique_words.add(word)

# Sort words
sorted_words = sorted(unique_words)

# Print results
print("\nUnique Words Analysis:")
print("Total Unique Words:", len(sorted_words))
print("Words:", sorted_words)
```
